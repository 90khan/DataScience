## Preprocessing of Data

### Train Test Split

```python
len(df)
```

```python
test_percent = 0.15
```

```python
len(df)*test_percent
```

```python
test_point = np.round(len(df)*test_percent) # we get 75% of the data, but since it's sequence, not random
test_point
```

```python
test_ind = int(len(df) - test_point) # Here we will use the 426th line before the training and after the test.
test_ind
```

### Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
train_scaled = scaler.fit_transform(train)
test_scaled = scaler.transform(test)
```

## Time Series Generator

```python
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
```

```python
#help(TimeseriesGenerator)
```

```python
# define generator
length = 2 # Length of the output sequences (in number of timesteps)
# length to train
batch_size = 1 # Number of timeseries samples in each batch
stride = 1 # Period between successive output sequences
generator = TimeseriesGenerator(data = train_scaled,
                                targets = train_scaled,
                                length = length,
                                batch_size = batch_size,
                                stride = stride)
# we make our data ready to give to the model
# get 3 rows guess 4th then get 3 rows (from 2nd row) guess 5th continue like this data boyo
# stride jhow many swipes?
```

## Modelling

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN, LSTM
```

```python
# We're only using one feature in our time series
n_features = 1
```

```python
seed = 42
```

```python
# define model
model = Sequential()

# Simple RNN layer
model.add(SimpleRNN(units = 100, activation = "tanh", input_shape=(length, n_features))) # input_shape (64.1) waiting for data

# Final Prediction
model.add(Dense(units = 1)) # 1 ouptut regression model
 
model.compile(optimizer='adam', loss='mse')
```

```python
model.summary()
```

```python
# params = 1 * ((size_of_input + 1) * size_of_output + size_of_output^2)
1 * ((1+1)*100+(100**2))
```

```python
model.fit_generator(generator = generator, epochs = 5)
```

```python
loss_df = pd.DataFrame(model.history.history)
loss_df.plot()
```

## Early Stopping and Validation Generator

```python
# We need to create our validation data to make an early stop
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor='val_loss', patience=2)
```

```python
length = 64 # lenght range have to be shorter than range of test data
            # In this case lenght range = 64, test range = 75
batch_size =1

generator = TimeseriesGenerator(data = train_scaled, targets = train_scaled, length = length, batch_size = batch_size)

validation_generator = TimeseriesGenerator(data = test_scaled, targets = test_scaled, length = length, batch_size = batch_size)
#the only difference is we give test_scaled as data and target
```

## LSTM

```python
# define model
model = Sequential()

# LSTM layer
model.add(LSTM(units = 100, activation = "tanh", input_shape = (length, n_features)))

# Final Prediction
model.add(Dense(units = 1))

model.compile(optimizer = 'adam', loss = 'mse')
```

```python
model.summary() # Parameter count increased 4 times in LSTM
```

```python
# params = 4 * ((size_of_input + 1) * size_of_output + size_of_output^2)
4 * ((1+1)*100+(100**2))
```

```python
model.fit_generator(generator = generator,
                    validation_data = validation_generator,
                    epochs = 20,
                    callbacks = [early_stop]) # for early_stop, must be validation_data
```

```python
loss_df = pd.DataFrame(model.history.history)
loss_df.plot()
```

## Retrain and Forecasting

```python
# we create our model with all the data
full_scaler = MinMaxScaler() 
scaled_full_data = full_scaler.fit_transform(df)
```

```python
length = 64 
generator = TimeseriesGenerator(scaled_full_data, scaled_full_data, length = length, batch_size = 1)
```

```python
model = Sequential()
model.add(LSTM(100, input_shape = (length, n_features)))
model.add(Dense(1))
model.compile(optimizer = 'adam', loss = 'mse')
model.fit_generator(generator, epochs = 6)
# 3 epochs are also trained so we did 6 epochs above so it wouldn't take long
# We got 6 epochs because I didn't have validation data and we couldn't make an early stop
# We make a forecast and add it to the end of the data and visualize it
```

```python
forecast = []

first_eval_batch = scaled_full_data[-length:]
current_batch = first_eval_batch.reshape((1, length, n_features))

for i in range(length):
    
    # get prediction 1 time stamp ahead 
    current_pred = model.predict(current_batch)
    
    # store prediction
    forecast.append(current_pred[0]) 
    
    # update batch to now include prediction and drop first value
    current_batch = np.append(current_batch[:, 1:, :], [current_pred], axis = 1)
```

```python
forecast = scaler.inverse_transform(forecast)
```
